####1. What is the difference between a function and a method in Python ?

A **function** is standalone and called directly (e.g., `func()`), while a **method** is tied to an object and called using dot notation (e.g., `obj.method()`). Functions exist globally, but methods are defined inside classes and can access the instance (`self`).

####2. Explain the concept of function arguments and parameters in Python.

Parameters are variables listed in a function's definition. They act as placeholders for the data the function will process.

````python
def greet(name):  # 'name' is a parameter
    return f"Hello, {name}!"
````

Arguments are the actual values passed to the function when it is called.


````python
print(greet("Alice"))  # "Alice" is an argument
````

####3. What are the different ways to define and call a function in Python?

###Defining Functions in Python

1. **Regular Function**:  
   Defined using the `def` keyword.  
   ```python
   def greet(name):
       return f"Hello, {name}!"
   ```

2. **Lambda Function**:  
   A single-expression anonymous function defined using `lambda`.  
   ```python
   greet = lambda name: f"Hello, {name}!"
   ```

---

###Calling Functions in Python

1. **Direct Call**:  
   Call the function by its name and pass required arguments.  
   ```python
   print(greet("Alice"))  # Outputs: Hello, Alice!
   ```

2. **As an Argument**:  
   Pass the function as an argument to another function.  
   ```python
   def executor(func, name):
       return func(name)

   print(executor(greet, "Bob"))  # Outputs: Hello, Bob!
   ```

3. **With Default Arguments**:  
   Call without providing arguments if default values are defined.  
   ```python
   def greet(name="Guest"):
       return f"Hello, {name}!"

   print(greet())  # Outputs: Hello, Guest!
   ```

4. **With Keyword Arguments**:  
   Specify arguments by name when calling.  
   ```python
   def greet(first, last):
       return f"Hello, {first} {last}!"

   print(greet(first="Alice", last="Smith"))  # Outputs: Hello, Alice Smith!
   ```

---

These are the most common ways to define and call functions in Python, allowing for flexibility in usage.

####5. What are iterators in Python and how do they differ from iterables ?

### **Iterators**:
- An **iterator** is an object that allows sequential access to its elements, one at a time, using the `next()` function.
- It implements the `__iter__()` and `__next__()` methods.
  
### **Iterables**:
- An **iterable** is any object that can be looped over (e.g., lists, strings, tuples) and can create an iterator using `iter()`.
- It must implement the `__iter__()` method.

### **Difference**:
- **Iterable**: Can be converted to an iterator (e.g., `[1, 2, 3]`).
- **Iterator**: Already provides one-item-at-a-time access. Once consumed, it can't be reused.

````python
numbers = iter([1, 2, 3])  # Create an iterator
print(next(numbers))
print(next(numbers))
````

####6. Explain the concept of generators in Python and how they are defined.

Generators in Python are a type of iterable, similar to lists or tuples, but they don't store all the values in memory at once. Instead, they produce items one at a time as they are needed, using a mechanism called lazy evaluation. This makes generators memory-efficient and particularly useful for handling large datasets or infinite sequences.

Generator Functions

1. Defined like a normal function using the def keyword.
2. Use the yield statement instead of return.

````python
# Defining a generator

def pow(n):
  for i in range(n):
    yield i**2
````

````python
pow(100)
````

````python
a = pow(100)
````

````python
next(a)
````

####7. What are the advantages of using generators over regular functions?

1. Generators are memory-efficient than regular functions because they don't store all the values in the memory at once.

2. They are particularly useful for handling large datasets or infinite sequences.

3. Since generators yield values on-demand and do not compute or store all values upfront, they can be faster for operations involving large data or streaming.

4. Generators simplify the creation of iterators. You don’t need to implement the entire iterator protocol (__iter__ and __next__ methods); you simply use yield.

5. Generators compute values only when they are requested, which reduces computational overhead.

####8. What is a lambda function in Python and when is it typically used?

A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike regular functions defined with def, lambda functions are concise and can have only a single expression, which is implicitly returned.

Syntax:
lambda arugments : expression

````python
# Using lambda with map()
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x ** 2, numbers))
print(squares)
````

####9. Explain the purpose and usage of the `map()` function in Python.

The map() function in Python is a built-in function used to apply a given function to each item of an iterable (e.g., list, tuple) and return an iterator containing the results. It is particularly useful for transforming data in a concise and efficient way without the need for explicit loops.

````python
str_numbers = ["1", "2", "3"]
int_numbers = map(int, str_numbers)
print(list(int_numbers))
````

####10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

The functions map(), reduce(), and filter() are built-in Python tools that are commonly used for functional programming. While they share similarities, each serves a unique purpose.

When to Use Each
1. Use map(): When you want to transform or modify every element in an iterable.

2. Use filter(): When you want to select elements based on a
condition.

3. Use reduce(): When you want to combine all elements into a single result.

````python
numbers = [1,2,3,4]
````

````python
result = map(lambda x: x ** 2, numbers)
print(list(result))
````

````python
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))
````

####4. Write a Python function that checks if a given number is prime or not from 1 to 200.

````python
def is_prime(n):
    if n < 2:
        return "Not a prime no."
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return "It is not a prime no."
    return "It is a prime no."
````

````python
is_prime(7)
````

#### 7. Implement a generator function that reads a file line by line and yields each line as a string.

````python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
````

####10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

````python
def vowel(n):
  vowels = 'aeiouAEIOU'
  return ''.join(filter(lambda x:x not in vowels,n))
````
